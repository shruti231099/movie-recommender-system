```python
m_cols=['movie_id','title',]
titles=pd.read_csv('u.item',sep='|',names=m_cols,usecols=range(2),encoding='latin-1')
```

```python
titles.head()
```

```python
x=ratings['user_id'].value_counts()>100
x[x].index
```

```python
y=x[x].index
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
data=pd.merge(ratings,titles,on='movie_id')
```

```python
data.shape
```

```python
number_rating=data.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'no_of_ratings'}, inplace=True)
```

```python
number_rating.head()
```

```python
final_data=data.merge(number_rating,on='title')
```

```python
final_data.shape
```

```python
final_data=final_data[final_data['no_of_ratings']>=50]
```

```python
final_data.shape
```

```python
movie_pivot=final_data.pivot_table(columns='user_id',index='title',values='rating')
```

```python
movie_pivot.fillna(0, inplace=True)
```

```python
movie_pivot
```

```python
from scipy.sparse import csr_matrix
movie_sparse=csr_matrix(movie_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(movie_sparse)
```

```python
def recommend_movie(movie_title):
    movie_id=np.where(movie_pivot.index==movie_title)[0][0]
    distances, suggestions=model.kneighbors(movie_pivot.iloc[movie_id, :].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestions)):
        print(movie_pivot.index[suggestions[i]])
```

```python
np.where(movie_pivot.index=='Godfather, The (1972)')[0][0]
```

```python
recommend_movie('Godfather, The (1972)')
```
